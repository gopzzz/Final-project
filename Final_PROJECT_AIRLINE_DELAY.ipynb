```python
#checking outliers
df.plot(kind='box', subplots=True, layout=(5,15),
sharex=False,sharey=False, figsize=(20,50), 
color='red')
```

```python
#Cleaning outliers
cols = ['DEP_DELAY','TAXI_OUT','TAXI_IN','ARR_DELAY','AIR_TIME','DISTANCE']

Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
df[cols]=np.where(df[cols]>upper_whisker,upper_whisker,np.where(df[cols]<lower_whisker,lower_whisker,df[cols]))
```

# **Modeling**

```python
X = df.drop(['DEP_DELAY','FL_DATE','DEP_TIME'], axis = 1)
# We are dropping the following values since 
# we believe they will not contribute much to the prediction
```

Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X)
X_scaled = scaler.transform(X)
X = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
y = df['DEP_DELAY']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

# **Random Forest Regressor**

```python
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mean_squared_error(y_test,y_pred)
```

Random Forest Regressor has produced an accuracy of 83.91

**Hyperparameter tuning using GridsearchCV**

```python
from sklearn.model_selection import GridSearchCV
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
}
grid = GridSearchCV(RandomForestRegressor() , param_grid , refit = True , verbose = 3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
grid.fit(X_train , y_train)
```

```python
grid.best_params_
```

**Forming a Random Forest model with the best parameters from GridSearchCV**

```python
model_rf = RandomForestRegressor(max_depth = 8, max_features = 'auto', n_estimators = 500)
model_rf.fit(X_train,y_train)
y_pred_RF_grid = model_rf.predict(X_test)
```

```python
model_rf.score(X_train,y_train)
```

```python
print('R-squared is: ',r2_score(y_test,y_pred_RF_grid))
```

**Hypertuning using RANDOMSEARCHCV**

```python
from sklearn.model_selection import RandomizedSearchCV
rf_RandomGrid = RandomizedSearchCV(estimator = model_rf, param_distributions = param_grid, cv = 10, verbose=2, n_jobs = 4)
```

```python
rf_RandomGrid.fit(X_train, y_train)
```

```python
rf_RandomGrid.best_params_
```

```python
rf_RandomGrid.score(X_train,y_train)
```

```python
random_y_pred = rf_RandomGrid.predict(X_test)
r2_score(y_test,random_y_pred)
```

With Hypertuning, we can see that the parameters can be selected as per requirement to tune the model and increase accuracy

# **Using XGBoost Regressor**

```python
import xgboost as xgb
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_squared_error
```

```python
xgbr = xgb.XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1,
       colsample_bynode=1, colsample_bytree=1, gamma=0,
       importance_type='gain', learning_rate=0.1, max_delta_step=0,
       max_depth=3, min_child_weight=1, missing=None, n_estimators=100,
       n_jobs=1, nthread=None, objective='reg:linear', random_state=0,
       reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
       silent=None, subsample=1, verbosity=1)
```

```python
xgbr.fit(X_train,y_train)
```

```python
score = xgbr.score(X_train,y_train)  
print("Training score: ", score)
```

Next, we can predict test data, then check the prediction accuracy. Here, we'll use MSE and RMSE as accuracy metrics.

```python
ypred = xgbr.predict(X_test)
print('R-squared is: ',r2_score(y_test,ypred))
```

Hyperparameter tuning of XBG using RandomCV

```python
params = { 'max_depth': [3, 5, 6, 10, 15, 20],
           'learning_rate': [0.01, 0.1, 0.2, 0.3],
           'subsample': np.arange(0.5, 1.0, 0.1),
           'colsample_bytree': np.arange(0.4, 1.0, 0.1),
           'colsample_bylevel': np.arange(0.4, 1.0, 0.1),
           'n_estimators': [100, 500, 1000]}
xgbr = xgb.XGBRegressor(seed = 20)
clf = RandomizedSearchCV(estimator=xgbr,
                         param_distributions=params,
                         scoring='neg_mean_squared_error',
                         n_iter=25,
                         verbose=1)
clf.fit(X_train, y_train)
print("Best parameters:", clf.best_params_)
```

```python
ypred_clf = clf.predict(X_test)
print('R-squared is: ',r2_score(y_test,ypred_clf))
```

**Conclusion:**
XGBoost model has the highest accuracy. Thus we can predict flight delay using this model.
